```python
import pandas as pd
from ucimlrepo import fetch_ucirepo

dow_jones_index = fetch_ucirepo(id=312)
df = dow_jones_index.data.features.copy()

# Here I'm cleaning up the data, namely removing the $ signs and converting to floats
price_cols = ["open", "high", "low", "close", "next_weeks_open", "next_weeks_close"]
for col in price_cols:
    df[col] = df[col].str.replace("$", "", regex=False).astype(float)

# Dropping missing values too
df["date"] = pd.to_datetime(df["date"])
df = df.dropna()

display(df.head())
print(df.dtypes)
```

Succesfully got rid of the $ signs, and data types changed to float as can be seen in this nice comparison

```python
# Here i export data into .csv
os.makedirs("exports", exist_ok=True)
df.to_csv("exports/dow_jones_raw.csv", index=False)
```
