```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
# Deciding which transactions are valid
valid = [
    t for t in transactions
    if t["price"] > 0 and t["quantity"] > 0
]

# Calculate totals
for t in valid:
    t["total"] = t["price"] * t["quantity"]

# Get top region and print
by_region = {}

for t in valid:
    region = t["region"]
    by_region[region] = by_region.get(region, 0) + t["total"]

top_region = max(by_region)
print(f"Top region: {top_region} | Revenue: {by_region[top_region]}")
```

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

def summarise_sales(data):
    total = sum(data)
    average = round(total/len(data), 2)

    return {
        "total": total,
        "average": average,
        "best_day": data.index(max(data)) + 1,
        "worst_day": data.index(min(data)) + 1,
        "above_average_count": len([x for x in data if x > average])
    }

result = summarise_sales(sales)
print(result)
```

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

passing = [s["name"] for s in students if s["grade"] >= 50]
print("Passing students", passing)

for s in students:
    name  = s["name"]
    grade = s["grade"]

    if grade >= 80:
        print(f"{name}: Excellent")
    elif grade >= 50:
        print(f"{name}: Pass")
    else:
        print(f"{name}: Fail")

# Attendance counter
low_attendance = 0
for s in students:
    if s["attendance"] < 0.80:
        low_attendance += 1

print("Low attendance count:", low_attendance)

# Finding student with highest grade (not using max())
highest_grade = students[0]
for s in students:
    if s["grade"] > highest_grade["grade"]:
        highest_grade = s

print(f"Best student (not max): {highest_grade['name']} | Grade: {highest_grade['grade']}")

# Finding highest grade using max()
def get_grade(s):
    return s["grade"]

top = max(students, key=get_grade)
print(f"Best student (max): {top['name']} | Grade: {top['grade']}")
```

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {"shares": shares, "price_per_share": price_per_share}

    def total_value(self):
        total = 0
        for d in self.holdings.values():
            total += d["shares"] * d["price_per_share"]
        return total

    def most_valuable(self):
        best_ticker = None
        best_value = 0
        for ticker, d in self.holdings.items():
            value = d["shares"] * d["price_per_share"]
            if value > best_value:
                best_value = value
                best_ticker = ticker
        return best_ticker

    #str method returning summary
    def __str__(self):
        return f"Owner: {self.owner} | Assets: {len(self.holdings)} | Total value: {self.total_value()}"


p = Portfolio("Axel")
p.add_asset("NOKI", 2, 150)
p.add_asset("ONIO", 6,  250)
p.add_asset("DOGG", 9,  420)

print(p.total_value())
print(p.most_valuable())
print(p)
```

```python
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
from datetime import datetime

# strptime reads the date according to the format i gave
dates = [datetime.strptime(order[0], "%Y-%m-%d").date() for order in orders]

# Subtracting gives the difference .days makes it look nice and clean
difference = (max(dates) - min(dates)).days

print(f"Earliest order: {min(dates)}")
print(f"Latest order:   {max(dates)}")
print(f"Days between:   {difference}")
```

```python
import numpy as np
# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])

# Annuals aka row sums
annual = revenue.sum(axis=1)
print("Annual revenue:", annual)

# Quarterlies i.e column means
quarterly_mean = revenue.mean(axis=0)
print("Quarterly mean:", quarterly_mean)

# Broadcasting
growth = np.array([1.05, 1.03, 1.08, 1.02])
adjustment = revenue[2] * growth
print("Adjusted 2024:", adjustment)

# Printing all above 250
print("Above 250:", revenue[revenue > 250])
```

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
# DataFrame w/date index
df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
}).set_index("date")

df["conversions"]    = round(df["visitors"] * df["conversion_rate"])
df["daily_revenue"]  = df["conversions"] * df["avg_order_value"]

print(df)
```
